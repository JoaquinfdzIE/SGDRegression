```python
import pandas as pd
```

```python
df = pd.read_pickle('final_dataset.pkl')
```

```python
df.head(-1)
```

```python
df = pd.get_dummies(df, columns=['event_type'], prefix='', prefix_sep='')
```

```python
df['view'] = df['view'].astype(int)
df['cart'] = df['cart'].astype(int)
df['purchase'] = df['purchase'].astype(int)
```

```python
df['year'] = df['year'].apply(lambda x: 1 if x == 2019 else 2)
```

```python
def assign_week_values(week_series):
    unique_weeks = week_series.unique()
    week_map = {week: idx for idx, week in enumerate(sorted(unique_weeks))}
    return week_series.map(week_map)
```

```python
df_new = df.copy()
```

```python
df_new['week'] = df_new.groupby('month')['week'].transform(assign_week_values)
```

```python
sample = df_new.sample(frac= 0.01, random_state=42)
```

```python
data = sample.to_pickle('muestra.pkl')
```
