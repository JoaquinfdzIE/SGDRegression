```python
# Import library
import pandas as pd
```

```python
# Read samples
df_march = pd.read_pickle('2020-Mar_sample.pkl')
```

```python
df_dec = pd.read_pickle('2019-Dec_sample.pkl')
```

```python
df_oct = pd.read_pickle('2019-Oct_sample.pkl')
```

```python
df_nov = pd.read_pickle('2019-Nov_sample.pkl')
```

```python
df_jan = pd.read_pickle('2020-Janu_sample.pkl')
```

```python
df_feb = pd.read_pickle('2020-Feb_sample.pkl')
```

```python
df_april = pd.read_pickle('2020-April_sample.pkl')
```

```python
# Join samples
df = pd.concat([df_oct, df_nov, df_dec, df_jan, df_feb, df_march, df_april], axis = 0)
```

```python
df.head(-1)
```

```python
# Creating new variables
df['day']=df['event_time'].dt.dayofweek
```

```python
df['week'] = df['event_time'].dt.isocalendar().week
```

```python
df['month'] = df['event_time'].dt.month
```

```python
df['year'] = df['event_time'].dt.year
```

```python
df.head(-1)
```

```python
# Export
data = df.to_pickle('final_dataset.pkl')
```
